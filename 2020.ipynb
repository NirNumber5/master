```python
import datetime
```

```python
def convert(val):
    try:
        i = val.index('.')
    except ValueError:
        j = val.index('+')
        val = val[:j]
        date_time_obj = datetime.datetime.strptime(val, '%Y-%m-%d %H:%M:%S')
    else:
        val = val[:i]
        date_time_obj = datetime.datetime.strptime(val, '%Y-%m-%d %H:%M:%S')
    return date_time_obj
```
